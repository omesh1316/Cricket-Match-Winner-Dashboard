# Cricket IPL Match Data - Exploratory Data Analysis

This notebook performs EDA on the IPL cricket matches dataset.

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Set style
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## 2. Load Dataset

```python
# Load the CSV file
df = pd.read_csv('../backend/data/matches.csv')

print("Dataset Shape:", df.shape)
print("\nFirst few rows:")
print(df.head(10))
```

## 3. Dataset Info

```python
print("Dataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nBasic Statistics:")
print(df.describe())
```

## 4. Team Analysis

```python
# Get unique teams
teams_team1 = df['team1'].unique()
teams_team2 = df['team2'].unique()
all_teams = set(teams_team1) | set(teams_team2)

print("Number of unique teams:", len(all_teams))
print("\nTeams:")
for team in sorted(all_teams):
    print(f"  - {team}")
```

## 5. Winner Distribution

```python
# Count wins by team
winner_counts = df['winner'].value_counts()

print("Matches won by each team:")
print(winner_counts)

# Plot
plt.figure(figsize=(12, 6))
winner_counts.plot(kind='bar', color='#10b981')
plt.title('Number of Matches Won by Each Team', fontsize=14, fontweight='bold')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 6. Toss Analysis

```python
# Toss winner vs match winner
toss_match_df = df[['toss_winner', 'winner']].copy()
toss_match_df['won_after_winning_toss'] = toss_match_df['toss_winner'] == toss_match_df['winner']

toss_wins = toss_match_df['won_after_winning_toss'].sum()
toss_loses = len(toss_match_df) - toss_wins

print(f"Matches won by toss winner: {toss_wins}")
print(f"Matches lost by toss winner: {toss_loses}")
print(f"Toss winner win percentage: {(toss_wins / len(toss_match_df)) * 100:.2f}%")

# Plot
toss_impact = pd.Series([toss_wins, toss_loses], index=['Won Match', 'Lost Match'])
plt.figure(figsize=(8, 6))
toss_impact.plot(kind='pie', autopct='%1.1f%%', colors=['#10b981', '#ef4444'])
plt.title('Toss Winner Match Outcome', fontsize=14, fontweight='bold')
plt.ylabel('')
plt.show()
```

## 7. Toss Decision Analysis

```python
# Bat vs Field decisions
decision_counts = df['toss_decision'].value_counts()

print("Toss Decisions:")
print(decision_counts)

# Plot
plt.figure(figsize=(8, 6))
decision_counts.plot(kind='bar', color=['#3b82f6', '#fb923c'])
plt.title('Toss Decision Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Decision')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## 8. Venue Analysis

```python
# Matches by venue
venue_counts = df['venue'].value_counts()

print(f"Number of unique venues: {df['venue'].nunique()}")
print("\nMatches by venue:")
print(venue_counts.head(10))

# Plot
plt.figure(figsize=(12, 6))
venue_counts.head(10).plot(kind='barh', color='#10b981')
plt.title('Top 10 Venues by Number of Matches', fontsize=14, fontweight='bold')
plt.xlabel('Number of Matches')
plt.tight_layout()
plt.show()
```

## 9. Data Quality Check

```python
# Check for missing target values
null_winners = df['winner'].isnull().sum()
print(f"Rows with null winner: {null_winners}")
print(f"Rows to use for training: {len(df) - null_winners}")

# Check data types
print("\nData types:")
print(df.dtypes)
```

## 10. Feature Summary

```python
# Summary of features used in the model
features = ['team1', 'team2', 'toss_winner', 'toss_decision', 'venue']

print("Feature Summary:")
print(f"Total Records: {len(df)}")
print(f"\nFeatures for Model:")

for feature in features:
    unique_count = df[feature].nunique()
    null_count = df[feature].isnull().sum()
    print(f"  - {feature}: {unique_count} unique values, {null_count} null values")

print(f"\nTarget Variable:")
print(f"  - winner: {df['winner'].nunique()} unique values")
```

## 11. Head-to-Head Analysis

```python
# Example: Mumbai Indians vs Chennai Super Kings
team1 = 'Mumbai Indians'
team2 = 'Chennai Super Kings'

h2h = df[((df['team1'] == team1) & (df['team2'] == team2)) | 
         ((df['team1'] == team2) & (df['team2'] == team1))]

print(f"Head-to-Head: {team1} vs {team2}")
print(f"Total matches: {len(h2h)}")

if len(h2h) > 0:
    team1_wins = (h2h['winner'] == team1).sum()
    team2_wins = (h2h['winner'] == team2).sum()
    print(f"{team1} wins: {team1_wins}")
    print(f"{team2} wins: {team2_wins}")
else:
    print("No matches found")
```

## Conclusion

This EDA provides insights into the IPL dataset structure and patterns. The data is ready for model training with all required features properly distributed.
